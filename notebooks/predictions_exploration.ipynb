```python
import os
import SimpleITK as sitk
import matplotlib.pyplot as plt
```

# Predictions exploration
This notebook inspects the predictions on the test set after having fully trained a model and evaluated it on the test set. Choose the model and the patient by setting the variables below:

```python
model_path = "../output/YOUR_MODEL/YOUR_RUN"
patient_id = 1      # 1-50
view = "2CH"        # 2CH, 4CH
ED_or_ES = "ED"     # ED, ES
```

Now run the cell below to display the predictions and the original echocardiogram.

```python
prediction = sitk.ReadImage(os.path.join(model_path, 'test_predictions', f"patient{patient_id:04d}_{view}_{ED_or_ES}.mhd"))
original = sitk.ReadImage(os.path.join("..", "data", "testing", f"patient{patient_id:04d}", f"patient{patient_id:04d}_{view}_{ED_or_ES}.mhd"))

# plot the images
plt.figure(figsize=(10, 15))
plt.subplot(1, 3, 1)
plt.imshow(sitk.GetArrayFromImage(original)[0], cmap='gray', vmin=0, vmax=255)
plt.subplot(1, 3, 2)
plt.imshow(sitk.GetArrayFromImage(prediction)[0])
plt.subplot(1, 3, 3)
plt.imshow(sitk.GetArrayFromImage(original)[0], cmap='gray', vmin=0, vmax=255)
plt.imshow(sitk.GetArrayFromImage(prediction)[0], alpha=0.5)
plt.show()
```
